### Подсчет простых статистик по сообщениям из чата ВК.

```python
YEAR = 2014
```

```python
import sys
sys.path.append("..")
from cozychat import CozyChat
```

```python
chat = CozyChat("мы приехали.txt")
```

## 1. Общая активность

Число сообщений за год.

```python
chat.show_total_messages_per_user(year=YEAR);
```

## 2. Активность по месяцам

```python
chat.show_total_messages_per_month(year=YEAR);
```

## 3. Активность по часам

```python
chat.show_total_messages_per_hour(year=YEAR);
```

## 4. Индивидуальная активность по часам

```python
chat.show_total_messages_per_hour_per_user(year=YEAR);
```

## 5. Топ популярных слов за год

Без учета "стоп-слов".

```python
chat.show_words_cloud(year=YEAR)
```

## 6. Индивидуальный топ популярных слов за год

```python
chat.show_words_cloud(year=YEAR, per_user=True)
```

## 7. Использование ненормативной лексики.

Число употреблений обсценной лексики по версии фильтра [отсюда](https://pypi.org/project/python-obscene-words-filter/).

```python
chat.show_obscene_usage(year=YEAR)
```

## 8. Попытки кого-нибудь куда-нибудь позвать.

Число употреблений слов из списка "призывных":
* "пойдем"
* "пошли",
* "пойдемте",
* "го",
* "хочет"

```python
chat.show_letsgo_usage(year=YEAR)
```

## 9. Рейтинг позитивности))).

Индекс позитивности - среднее число непарных закрывающих скобок `)` в сообщении. Рейтинг позитивности подзаголовка выше = `3.0`

```python
chat.show_positive_usage(year=YEAR)
```

## 9. Эмоциональность сообщений.

Доля сообщений (от числа своих же) с ярко выраженной эмоциональной окраской по версии фильтра из библиотеки [`dostoevsky`](https://github.com/bureaucratic-labs/dostoevsky).

```python
chat.show_sentiment(year=YEAR)
```
